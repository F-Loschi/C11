##Importando bibliotecas e csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('paises.csv', delimiter=";")
```

##Exercício 1: Por meio do dataset paises.csv, trace dois gráficos de linhas em um mesmo plano cartesiano, um mostrando a taxa de mortalidade (Deathrate) e outro a taxa de natalidade (Birthrate) dos países da América do Norte;

```python
df
```

```python
df_america = df[df['Region'].str.contains('NORTHERN AMERICA')]
df_america
```

```python

deathrate = df_america['Deathrate']
birthrate = df_america['Birthrate']
y = df_america['Country']

plt.plot(deathrate, y, '*-r', linewidth=3, markersize=10, label='Deathrate')
plt.plot(birthrate, y, 's-g', linewidth=3, markersize=10, label='Birthrate')

plt.xlabel("Taxa (%)")
plt.ylabel("Países da América do Norte")
plt.title("Taxa de Mortalidade e Natalidade")
plt.legend()
plt.grid(True)

plt.show()
```

## Exercício 3: Por meio do dataset space.csv, trace um gráfico em torta ilustrando a porcentagem de missões da empresa Roscosmos que deram certo e que deram errado;

```python
certo = df2[(df2['Company Name']== 'Roscosmos')&(df2['Status Mission'] == 'Success')]
errado = df2[(df2['Company Name']== 'Roscosmos')&(df2['Status Mission'] == 'Failure')]

qtCerto = len(certo)
qtErrado = len(errado)
```

```python
plt.pie([qtCerto, qtErrado], labels=['Certo', 'Errado'], autopct='%1.1f%%')
plt.title('Porcentagem de Missões da Empresa Roscosmos')
plt.show()
```
